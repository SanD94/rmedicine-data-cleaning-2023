```r
library(tidyverse)
library(here)

excel_file = here("data", "messy_uc.xlsx")
df_raw <- readxl::read_excel(excel_file, sheet = "Data", skip = 5) |>
  filter(!is.na(pat_id)) |>
  slice_head(n=10) |>
  select(pat_id:dob, start_bp, start_mes)


df_raw
```

```r
display_html_jupyter <- function(tbl) {
    tbl |>
        capture.output() |>
        paste(collapse = '') |>
        IRdisplay::display_html()
}

display_md_jupyter <- function(tbl) {
    tbl |>
        capture.output() |>
        paste(collapse = '') |>
        IRdisplay::display_markdown()
}
```
